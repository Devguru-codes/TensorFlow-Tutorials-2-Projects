## Q5 - Recreate Softmax function

```python
import numpy as np
import tensorflow as tf

def softmax(A):
    A_max = tf.reduce_max(A)  # Get max value for numerical stability
    exp_A = tf.exp(A - A_max)  # Subtract max before exponentiation
    return exp_A / tf.reduce_sum(exp_A)  # Normalize

A = np.arange(10, 35, 2)  # Array from 10 to 34 with step 2
A = tf.cast(A, dtype=tf.float32)  # Use float32 to avoid precision loss
softmax_values = softmax(A)

print(softmax_values)
```

## Q7 - Multiclass MNIST model

```python
(X_train, train_labels), (y_test, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
X_train, y_test = X_train / 255.0, y_test / 255.0
```

```python
#Set seed
tf.random.set_seed(42)
#1.Create the model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')
])
#2.Compile it
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)
#3.Fit it
history = model.fit(X_train,train_labels,epochs=10)
```

```python
import random
import matplotlib.pyplot as plt

def plot_visualise(size,data,labels):
  i = random.randint(0,size-1)
  pred = model.predict(data[i].reshape(1, 28, 28))
  pred = pred.argmax(axis = 1)
  pred_class = pred[0]
  plt.figure(figsize=(6, 6))  # Increase figure size
  plt.imshow(data[i],cmap = plt.cm.binary)
  plt.title(f"truth:{class_names[labels[i]]} vs pred:{class_names[pred_class]}")
  plt.axis(False)
  plt.tight_layout() # Adjust layout
  plt.show()
```

```python
data = np.concatenate((X_train, y_test))
labels = np.concatenate((train_labels, test_labels))
len(data),len(labels)
#Create a small list so we can index onto our training labels so they are human readable
class_names = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
len(class_names)
```

```python
plot_visualise(len(data),data,labels)
```
