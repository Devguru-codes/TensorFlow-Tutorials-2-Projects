#shuffle order of elements in a tensor
* change order so that inherent data doesn't effect learning

```python
not_shuff = tf.constant([[10,7],
                         [3,50],
                          [3,7]])
not_shuff.ndim
```

```python
not_shuff
```

```python
tf.random.set_seed(42)
shuff = tf.random.shuffle(not_shuff,seed =42)
shuff
```

```python
tf.random.set_seed(42) #-> global level seed
shuff = tf.random.shuffle(not_shuff,seed =40) #->operational level seed
shuff
```

```python
tf.random.set_seed(1234)
t1 = tf.random.Generator.from_seed(1234)
t1 = t1.normal(shape=(2,3))
t1
```

```python
tf.random.set_seed(1234)
t2 = tf.random.shuffle(t1,seed =42)
t2
```

```python
#tensor of same shape
tf1 = tf.random.normal(shape = (3,2,3))
tf2 = tf.random.normal(shape = (3,2,3))
tf.matmul(tf1,tf2)
```

rules:-
* the inner dimensions must match
* resulting matrix must have same shape of outer dimensions

```python
#transpose
tf1,tf.transpose(tf1)
```

```python
X = tf.random.normal(shape = (3,2))
Y = tf.random.normal(shape = (3,2))
Z = tf.tensordot(X,tf.transpose(Y),axes = 1)
W = tf.tensordot(X,tf.reshape(Y,shape = (2,3)),axes = 1)
comparison_result = tf.reduce_all(tf.equal(Z, W))
if(comparison_result):
  print("True")
else:
  print("false")
```

```python
print("Y",Y)
print("transpose:",tf.transpose(Y))
print("reshape:",tf.reshape(Y,shape = (2,3)))
```

```python
# create a random tensor with values b/w 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100,size = 50))
E
```

```python
tf.reduce_mean(E)
```

```python
tf.reduce_sum(E)
```

```python
variance = tf.math.reduce_variance(tf.cast(E, tf.float32)) # Cast to float32 for variance calculation
print("Variance:", variance.numpy())

std_dev = tf.math.reduce_std(tf.cast(E, tf.float32)) # Cast to float32 for std_dev calculation
print("Standard Deviation:", std_dev.numpy())
```

# Positional maximum and minimum

```python
#create a new tensor
tf.random.set_seed(42)
F = tf.random.uniform(shape = [50])
F
```

```python
#find positional max
tf.argmax(F)
```

```python
F[tf.argmax(F)]
```

```python
tf.reduce_max(F)
```

```python
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
#find positional min
tf.argmin(F)
```

```python
F[tf.argmin(F)]
```

```python
tf.reduce_min(F)
```

```python
F[tf.argmin(F)] == tf.reduce_min(F)
```
