```python
import tensorflow as tf
```

```python
sc = tf.constant(1)
vec = tf.constant([1,2],shape = [2])
mat = tf.constant([[1,2],[3,4]],shape = [2,2])
ten = tf.random.uniform(shape = [2,3,4])
print(sc,vec,mat,ten)
```

```python
print(sc.shape,sc.ndim,tf.size(sc))
print(vec.shape,vec.ndim,tf.size(vec))
print(mat.shape,mat.ndim,tf.size(mat))
print(ten.shape,ten.ndim,tf.size(ten))
```

```python
tf.random.set_seed(42)
t1 = tf.random.uniform(shape = [5, 300])
t2 = tf.random.uniform(shape = [5, 300])
print(t1,t2)
```

```python
tf.matmul(t1,tf.transpose(t2))
```

```python
tf.tensordot(t1,tf.transpose(t2),axes = 1)
```

```python
t3 = tf.random.uniform(shape = [224, 224, 3])
t3
```

```python
print(tf.reduce_min(t3,axis = 1))
print(tf.reduce_max(t3,axis = 1))
```

```python
t4 = tf.random.uniform(shape = [1, 224, 224, 3])
t4
```

```python
sq_t4 = tf.squeeze(t4)
sq_t4
```

```python
t5 = tf.random.normal(shape = [10])
t5
```

```python
print(tf.argmin(t5))
print(tf.argmax(t5))
```

```python
tf.one_hot(tf.cast(t5,dtype = tf.int32),depth = 10)
```
