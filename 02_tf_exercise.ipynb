```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(-1000,2000,2.5)
y = np.arange(-2000,4000,5)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
X_train = tf.constant(X_train,shape=(960,1))
X_test = tf.constant(X_test,shape=(240,1))
y_train = tf.constant(y_train,shape=(960,1))
y_test = tf.constant(y_test,shape=(240,1))

```

```python
tf.random.set_seed(42)
#1.Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#2.compile it

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

model.fit(tf.expand_dims(X_train,-1),y_train,epochs = 100)
```

```python
tf.random.set_seed(42)
#1.Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
])

#2.compile it

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

model.fit(tf.expand_dims(X_train,-1),y_train,epochs = 200)
```

```python
model.evaluate(tf.expand_dims(X_test,-1),y_test)
```

```python
tf.random.set_seed(42)
#1.Create a model
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#2.compile it

model1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

model1.fit(tf.expand_dims(X_train,-1),y_train,epochs = 200)
```

```python
model1.evaluate(tf.expand_dims(X_test,-1),y_test)
```

```python
df = tf.keras.datasets.boston_housing.load_data()
```

```python
import pandas as pd

# Assuming 'df' is the output of tf.keras.datasets.boston_housing.load_data()
(x_train, y_train), (x_test, y_test) = df

# Create pandas DataFrames
train_df = pd.DataFrame(x_train, columns=[f"feature_{i}" for i in range(x_train.shape[1])])


test_df = pd.DataFrame(x_test, columns=[f"feature_{i}" for i in range(x_test.shape[1])])


# Now you have two pandas DataFrames: train_df and test_df
print(train_df.head())
print(test_df.head())
```

```python
tf.random.set_seed(42)
#1.Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#2.compile it

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

history = model.fit(train_df,y_train,epochs = 200)
```

```python
model.evaluate(test_df,y_test)
```

```python
tf.random.set_seed(42)
#1.Create a model
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
])

#2.compile it

model_1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

model_1.fit(train_df,y_train,epochs = 200)
```

```python
model_1.evaluate(test_df,y_test)
```

```python
tf.random.set_seed(42)
#1.Create a model
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#2.compile it

model1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001),
    metrics = ["mae"]
)
#3.fit it

model1.fit(train_df,y_train,epochs = 200)
```

```python
model1.evaluate(test_df,y_test)
```
