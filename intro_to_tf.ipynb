# **rank or degree of tensor**

```python
rank1_tensor = tf.Variable(["test"],tf.string)
rank2_tensor = tf.Variable([["test","ok","good"],["i","am","batman"]],tf.string)
print("1st shape:",rank1_tensor.shape)
print("1st rank:",tf.rank(rank1_tensor))
print("2nd shape:",rank2_tensor.shape)
print("2nd rank:",tf.rank(rank2_tensor))
```

# **Changing shape**
we use tf.reshape for this. now number of elements must be same for reshape and multiplication of dimensions.

```python
tensor1 = tf.ones([1,2,3])
tensor2 = tf.reshape(tensor1,[2,3,1])
tensor3 = tf.reshape(tensor2,[3,-1])
#using reshape with -1 allows you to specify some dimensions while letting TensorFlow automatically
#calculate the others, ensuring that all the elements from the original tensor are preserved in the reshaped
#tensor here it knew 6 are total elements 3 rows are needed and it calculated no of cols by 6/3 = 2 cols.
print(tensor1)
print(tensor2)
print(tensor3)
```
